```python
import matplotlib.pyplot as plt
import pandas as pd

file = "../SimulationData/Thea/Timeblocks100000Steps64Walldensity1-5Fluiddensity0-7_1.dat"

table = pd.read_csv(file,delimiter=" ",header=None,names=["pos_z","density","velocity","temperature"])

fig,ax = plt.subplots(1,3,sharex=False,figsize=(15,6))
ax[0].plot(table["pos_z"],table["density"])
ax[0].set_title("density")

ax[1].plot(table["pos_z"],table["velocity"])
ax[1].set_title("velocity")

ax[2].plot(table["pos_z"],table["temperature"])
ax[2].set_title("temperature")
```
