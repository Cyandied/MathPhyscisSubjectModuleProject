```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

file = "../SimulationData/Thea/Timeblocks15000Steps64_1.dat"

db = pd.read_csv(file,delimiter=" ",header=None,names=["pos_z","density","velocity","temperature"])

fig,ax = plt.subplots(1,3,sharex=False,figsize=(15,6))
ax[0].plot(db["pos_z"],db["density"])
ax[0].set_title("density")

ax[1].plot(db["pos_z"],db["velocity"])
ax[1].set_title("velocity")

ax[2].plot(db["pos_z"],db["temperature"])
ax[2].set_title("temperature")
```
